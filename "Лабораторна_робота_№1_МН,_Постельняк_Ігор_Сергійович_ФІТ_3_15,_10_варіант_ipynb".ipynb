```python
# Постельняк Ігор Сергійович ФІТ 3-15
# Варіант 10

# Завдання 1
# 10. Створіть одновимірний масив (вектор) з перших 10 піднесених до квадрату натуральних чисел та виведіть його значення.
import numpy as np
vector10 = np.array([i**2 for i in range(1, 11)])
print("Завдання 1.10:", vector10)
```

```python
# Завдання 2
# 10. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями квадратних коренів від кожного натурального числа, починаючи з 1, та виведіть його значення.
import numpy as np
import math
matrix10 = np.array([[math.sqrt(i*3 + j + 1) for j in range(3)] for i in range(3)])
print("Завдання 2.10:")
print(matrix10)
```

```python
# Завдання 3
# 10. Створіть масив розміром 3x3, заповніть його випадковими цілими числами в діапазоні від 0 до 5 та виведіть його значення.
import numpy as np
matrix10 = np.random.randint(0, 6, (3, 3))
print("Завдання 3.10:")
print(matrix10)
```

```python
# Завдання 4
# 10. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть суму елементів, які менше середнього значення.
import numpy as np
matrix = np.random.rand(5, 5)
mean_val = np.mean(matrix)
sum_less = np.sum(matrix[matrix < mean_val])
print("Завдання 4.10:")
print("Матриця:")
print(matrix)
print(f"Середнє: {mean_val:.4f}")
print(f"Сума елементів менших за середнє: {sum_less:.4f}")
```

```python
# Завдання 5
# 10. Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими числами від 1 до 30 та виконайте на них поелементні операції додавання, віднімання та множення.
import numpy as np
arr1 = np.random.randint(1, 31, 5)
arr2 = np.random.randint(1, 31, 5)
print("Завдання 5.10:")
print("Масив 1:", arr1)
print("Масив 2:", arr2)
print("Додавання:", arr1 + arr2)
print("Віднімання:", arr1 - arr2)
print("Множення:", arr1 * arr2)
```

```python
# Завдання 6
# Створіть дві матриці 3x3, заповніть їх довільними дійсними числами від 0 до 1 та знайдіть їх добуток.
import numpy as np
matrix1 = np.random.rand(3, 3)
matrix2 = np.random.rand(3, 3)
product = np.dot(matrix1, matrix2)
print("Завдання 6.10:")
print("Матриця 1:")
print(matrix1)
print("Матриця 2:")
print(matrix2)
print("Добуток матриць:")
print(product)
```

```python
# Завдання 7
# Створіть дві матриці розміром 2x3 та 3x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
import numpy as np
matrix1 = np.random.randint(1, 6, (2, 3))
matrix2 = np.random.randint(1, 6, (3, 2))
product = np.dot(matrix1, matrix2)
print("Завдання 7.10:")
print("Матриця 2x3:")
print(matrix1)
print("Матриця 3x2:")
print(matrix2)
print("Добуток:")
print(product)
```

```python
# Завдання 8
# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю.
import numpy as np
matrix = np.random.randint(1, 11, (3, 3))
try:
    inv_matrix = np.linalg.inv(matrix)
    print("Завдання 8.10:")
    print("Початкова матриця:")
    print(matrix)
    print("Обернена матриця:")
    print(inv_matrix)
except np.linalg.LinAlgError:
    print("Матриця вироджена, оберненої не існує")
```

```python
# Завдання 9
# Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.
import numpy as np
matrix = np.random.rand(5, 5)
transposed = matrix.T
print("Завдання 9.10:")
print("Початкова матриця:")
print(matrix)
print("Транспонована матриця:")
print(transposed)
```

```python
# Завдання 10
# Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими цілими числами в діапазоні від -5 до 5, а вектор - випадковими дійсними числами від 0 до 1 і знайдіть добуток матриці на вектор.
import numpy as np
matrix = np.random.randint(-5, 6, (3, 4))
vector = np.random.rand(4, 1)
result = np.dot(matrix, vector)
print("Завдання 10.10:")
print("Матриця 3x4:")
print(matrix)
print("Вектор 4x1:")
print(vector)
print("Добуток матриці на вектор:")
print(result)
```

```python
# Завдання 11
# Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток матриці на вектор.
import numpy as np
matrix = np.random.rand(2, 3)
vector = np.random.rand(3, 1)
result = np.dot(matrix, vector)
print("Завдання 11.10:")
print("Матриця 2x3:")
print(matrix)
print("Вектор 3x1:")
print(vector)
print("Добуток матриці на вектор:")
print(result)
```

```python
# Завдання 12
# Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть суму всіх елементів обох матриць.
import numpy as np
matrix1 = np.random.randint(1, 11, (3, 3))
matrix2 = np.random.randint(1, 11, (3, 3))
total_sum = np.sum(matrix1) + np.sum(matrix2)
print("Завдання 12.10:")
print("Матриця 1:")
print(matrix1)
print(f"Сума елементів матриці 1: {np.sum(matrix1)}")
print("Матриця 2:")
print(matrix2)
print(f"Сума елементів матриці 2: {np.sum(matrix2)}")
print(f"Загальна сума: {total_sum}")
```

```python
# Завдання 13
# Створіть дві матриці розміром 2x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть різницю їх елементів.
import numpy as np
matrix1 = np.random.randint(1, 11, (2, 2))
matrix2 = np.random.randint(1, 11, (2, 2))
difference = matrix1 - matrix2
print("Завдання 13.10:")
print("Матриця 1:")
print(matrix1)
print("Матриця 2:")
print(matrix2)
print("Різниця (матриця1 - матриця2):")
print(difference)
```

```python
# Завдання 14
# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 5 та знайдіть кількість парних елементів матриці.
import numpy as np
matrix = np.random.randint(1, 6, (3, 3))
even_count = np.sum(matrix % 2 == 0)
print("Завдання 14.10:")
print("Матриця 3x3:")
print(matrix)
print(f"Кількість парних елементів: {even_count}")
```

```python
# Завдання 15
# Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму кубів елементів кожного рядка матриці.
import numpy as np
matrix = np.random.uniform(-1, 1, (4, 4))
cubes_sum = np.sum(matrix**3, axis=1).reshape(-1, 1)
print("Завдання 15.10:")
print("Матриця 4x4:")
print(matrix)
print("Вектор-стовпчик сум кубів елементів кожного рядка:")
print(cubes_sum)
```
